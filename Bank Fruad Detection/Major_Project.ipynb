````python
# Importing the necessary libraries needed
import pandas as pd
````

````python
# reading the dataset present in csv
df = pd.read_csv('/content/bank_fraud_dataset.csv')
````

````python
# Droping the index column 
df.drop('Unnamed: 0', inplace=True, axis = 1)
````

# Description: 
 Since there is no missing data or nan values so no need to Clean the data<br>

>For X ▶ the attribute pcirle has only one value for entire dataset so no need to consider that attribute and exclude label



> For Y ▶ We Consider the label only 






````python
X = df.drop(df[['pcircle', 'label']], axis=1)
y = df['label']
````

The datatype of the dataset contains both numeric columns and text columns <br>so we need to encode the text to numeric value using LabelEncoder

The Text columns can be seperated by 
```
X.select_dtypes(include='object').columns
```



````python
# Importing LabelEncoder 
from sklearn.preprocessing import LabelEncoder

lr = LabelEncoder()
for i in X.select_dtypes(include='object').columns:
  X[i] = lr.fit_transform(X[i])
````

Spliting the dataset for training and testing using train_test_split in sklearn


1.   training data - 80%
2.   testing data - 20%



````python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
````

````python
# Importing the Classification Algorithm 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
````

````python
# Creating a dictionaries with the name as keys
pipeline = {'KNN' : KNeighborsClassifier(),
            'AdaBoostClassifier' : AdaBoostClassifier(),
            'RandomForestClassifier' : RandomForestClassifier(),
            'Naive Bayes' : GaussianNB()}
````

````python
# Importing accuracy_score, f1_score,confusion_matrix to check how good our model performs
from sklearn.metrics import accuracy_score, f1_score,confusion_matrix

#Iterating through the various Algorithm 
for alg in pipeline:
  pipeline[alg].fit(X_train, Y_train)
  ac = accuracy_score(pipeline[alg].predict(X_test), Y_test) * 100
  fn = f1_score(pipeline[alg].predict(X_test), Y_test) * 100
  print(f'\nusing {alg} Algorithm : ')
  print('\tAccuracy_score: %.2f %%\n\tF1_score :  %.2f %%' %(ac, fn))
````

As you can see in above observation **Random Forest Classifier** had performed the best <br> so we train our model with whole dataset 

````python
trained_model = pipeline['RandomForestClassifier'].fit(X, y)
````

We trained our whole dataset with **Random Forest Classifier.**<br>Hence we can save our trained model and use it for further purpose

````python
# Importing picle module to save the trained model
import pickle

with open('trained_model', 'wb') as model:
  pickle.dump(trained_model, model)
````
